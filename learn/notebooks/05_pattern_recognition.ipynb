## Pattern Skeleton Code

```python
# Two Pointers skeleton
def two_pointers(arr: list[int], target: int) -> list[int]:
    """Two pointers from both ends."""
    left, right = 0, len(arr) - 1
    while left < right:
        current = arr[left] + arr[right]
        if current == target:
            return [left, right]
        elif current < target:
            left += 1
        else:
            right -= 1
    return []

# Sliding Window skeleton
def sliding_window(arr: list[int], k: int) -> int:
    """Fixed size sliding window."""
    window_sum = sum(arr[:k])
    max_sum = window_sum
    
    for i in range(k, len(arr)):
        window_sum += arr[i] - arr[i - k]  # Add new, remove old
        max_sum = max(max_sum, window_sum)
    
    return max_sum
```

```python
# Fast & Slow Pointers skeleton
class ListNode:
    def __init__(self, val: int = 0, next: 'ListNode | None' = None):
        self.val = val
        self.next = next

def has_cycle(head: ListNode | None) -> bool:
    """Detect cycle using fast and slow pointers."""
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False

# Binary Search skeleton
def binary_search(arr: list[int], target: int) -> int:
    """Standard binary search."""
    left, right = 0, len(arr) - 1
    
    while left <= right:
        mid = left + (right - left) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return -1
```

## Key Takeaways

### Pattern Recognition Checklist:
1. **Read the problem carefully** - identify input types and constraints
2. **Look for keywords** - sorted, substring, intervals, cycle, etc.
3. **Check constraints** - time/space requirements hint at patterns
4. **Consider trade-offs** - some patterns can be combined

### Quick Pattern Selection:
- **Sorted array** → Two Pointers or Binary Search
- **Subarray/substring** → Sliding Window
- **Linked list** → Fast & Slow Pointers
- **Intervals** → Merge Intervals
- **1 to n numbers** → Cyclic Sort
- **Tree levels** → Tree BFS
- **Tree paths** → Tree DFS
- **All combinations** → Subsets
- **K elements** → Top K / Heap
- **Optimization** → Dynamic Programming
