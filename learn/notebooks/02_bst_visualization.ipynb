# 🌳 Binary Search Tree Visualization

Interactive exploration of BST operations.

```python
from typing import Optional
import matplotlib.pyplot as plt
import networkx as nx
from collections import deque
```

## BST Implementation

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left: Optional['TreeNode'] = None
        self.right: Optional['TreeNode'] = None

class BST:
    def __init__(self):
        self.root: Optional[TreeNode] = None
    
    def insert(self, val):
        if not self.root:
            self.root = TreeNode(val)
            return
        self._insert(self.root, val)
    
    def _insert(self, node, val):
        if val < node.val:
            if node.left is None:
                node.left = TreeNode(val)
            else:
                self._insert(node.left, val)
        else:
            if node.right is None:
                node.right = TreeNode(val)
            else:
                self._insert(node.right, val)
    
    def inorder(self):
        result = []
        self._inorder(self.root, result)
        return result
    
    def _inorder(self, node, result):
        if node:
            self._inorder(node.left, result)
            result.append(node.val)
            self._inorder(node.right, result)
    
    def to_graph(self):
        G = nx.DiGraph()
        if not self.root:
            return G
        queue = deque([(self.root, None)])
        while queue:
            node, parent = queue.popleft()
            G.add_node(node.val)
            if parent:
                G.add_edge(parent.val, node.val)
            if node.left:
                queue.append((node.left, node))
            if node.right:
                queue.append((node.right, node))
        return G
```

## Build and Visualize BST

```python
values = [50, 30, 70, 20, 40, 60, 80, 10, 25, 35, 45]
bst = BST()
for v in values:
    bst.insert(v)

print(f"Inserted values: {values}")
print(f"Inorder traversal: {bst.inorder()}")
```

```python
G = bst.to_graph()
pos = nx.nx_agraph.graphviz_layout(G, prog='dot') if False else nx.spring_layout(G, k=2, iterations=50)

plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, node_color='lightblue', 
        node_size=1500, font_size=12, font_weight='bold',
        arrows=True, edge_color='gray')
plt.title('Binary Search Tree Visualization')
plt.show()
```

## BST Properties

- **Left child** < parent < **Right child**
- **Inorder traversal** gives sorted order
- **Search**: O(log n) average, O(n) worst case
- **Insert/Delete**: O(log n) average
