# Graph Traversal Visualization

Step-by-step visualization of BFS (Breadth-First Search) and DFS (Depth-First Search) algorithms.

**OCTALUM-PYLAB** - Master Python Data Structures & Algorithms

```python
import matplotlib.pyplot as plt
import numpy as np
from collections import deque

%matplotlib inline
```

## Part 1: Building the Graph

Using adjacency list representation - the most common Python graph structure.

```python
# Sample graph as adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# Node positions for visualization
positions = {
    'A': (0, 2),
    'B': (-1, 1),
    'C': (1, 1),
    'D': (-1.5, 0),
    'E': (-0.5, 0),
    'F': (1, 0)
}
```

```python
def draw_graph(ax, graph: dict, positions: dict, visited: set, current: str | None = None, 
               queue: list | None = None, title: str = "Graph"):
    """Draw graph with visited nodes highlighted."""
    ax.clear()
    ax.set_title(title, fontsize=12, fontweight='bold')
    
    # Draw edges
    for node, neighbors in graph.items():
        for neighbor in neighbors:
            if node < neighbor:  # Avoid drawing edges twice
                x1, y1 = positions[node]
                x2, y2 = positions[neighbor]
                ax.plot([x1, x2], [y1, y2], 'gray', linewidth=2, alpha=0.5)
    
    # Draw nodes
    for node, (x, y) in positions.items():
        if node == current:
            color = '#e74c3c'  # Red for current
            size = 800
        elif node in visited:
            color = '#27ae60'  # Green for visited
            size = 600
        else:
            color = '#3498db'  # Blue for unvisited
            size = 600
        
        circle = plt.Circle((x, y), 0.2, color=color, ec='white', linewidth=2)
        ax.add_patch(circle)
        ax.text(x, y, node, ha='center', va='center', fontsize=14, fontweight='bold', color='white')
    
    # Show queue/stack if provided
    if queue:
        ax.text(0, -1, f'Queue/Stack: {queue}', ha='center', fontsize=10, 
                bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    
    ax.set_xlim(-2.5, 2.5)
    ax.set_ylim(-1.5, 3)
    ax.set_aspect('equal')
    ax.axis('off')
```

```python
# Draw initial graph
fig, ax = plt.subplots(figsize=(10, 6))
draw_graph(ax, graph, positions, set(), title="Sample Graph (Adjacency List)")
plt.show()

print("Adjacency List Representation:")
for node, neighbors in sorted(graph.items()):
    print(f"  {node}: {neighbors}")
```

## Part 2: Breadth-First Search (BFS)

- Uses a **queue** (FIFO)
- Explores level by level
- Guarantees shortest path in unweighted graphs
- Time: O(V + E), Space: O(V)

```python
def bfs_with_steps(graph: dict, start: str) -> list[tuple[set, str | None, list]]:
    """BFS that records each step for visualization."""
    steps = []
    visited = set()
    queue = deque([start])
    
    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.add(node)
            steps.append((visited.copy(), node, list(queue)))
            
            for neighbor in graph[node]:
                if neighbor not in visited and neighbor not in queue:
                    queue.append(neighbor)
    
    return steps
```

```python
# Run BFS and visualize steps
bfs_steps = bfs_with_steps(graph, 'A')

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i, (visited, current, queue) in enumerate(bfs_steps):
    draw_graph(axes[i], graph, positions, visited, current, queue, 
               title=f"BFS Step {i+1}: Processing {current}")

# Hide unused subplots
for i in range(len(bfs_steps), len(axes)):
    axes[i].axis('off')

plt.tight_layout()
plt.show()

print("\nBFS Order:", ' -> '.join([step[1] for step in bfs_steps]))
```

## Part 3: Depth-First Search (DFS)

- Uses a **stack** (or recursion) - LIFO
- Explores as deep as possible before backtracking
- Does NOT guarantee shortest path
- Time: O(V + E), Space: O(V)

```python
def dfs_with_steps(graph: dict, start: str) -> list[tuple[set, str | None, list]]:
    """DFS that records each step for visualization."""
    steps = []
    visited = set()
    stack = [start]
    
    while stack:
        node = stack.pop()
        if node not in visited:
            visited.add(node)
            steps.append((visited.copy(), node, list(stack)))
            
            # Add neighbors in reverse order for consistent traversal
            for neighbor in reversed(graph[node]):
                if neighbor not in visited:
                    stack.append(neighbor)
    
    return steps
```

```python
# Run DFS and visualize steps
dfs_steps = dfs_with_steps(graph, 'A')

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i, (visited, current, stack) in enumerate(dfs_steps):
    draw_graph(axes[i], graph, positions, visited, current, stack, 
               title=f"DFS Step {i+1}: Processing {current}")

# Hide unused subplots
for i in range(len(dfs_steps), len(axes)):
    axes[i].axis('off')

plt.tight_layout()
plt.show()

print("\nDFS Order:", ' -> '.join([step[1] for step in dfs_steps]))
```

## Part 4: BFS vs DFS Comparison

```python
# Side-by-side comparison
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Final BFS state
bfs_visited = bfs_steps[-1][0] if bfs_steps else set()
draw_graph(axes[0], graph, positions, bfs_visited, 
           title=f"BFS Final Order: {' -> '.join([s[1] for s in bfs_steps])}")

# Final DFS state
dfs_visited = dfs_steps[-1][0] if dfs_steps else set()
draw_graph(axes[1], graph, positions, dfs_visited,
           title=f"DFS Final Order: {' -> '.join([s[1] for s in dfs_steps])}")

plt.tight_layout()
plt.show()
```

## Key Takeaways

### When to Use BFS:
- Finding shortest path in unweighted graphs
- Level-order traversal (e.g., organizational hierarchy)
- Finding all nodes within K distance
- Peer-to-peer networks

### When to Use DFS:
- Topological sorting
- Detecting cycles in a graph
- Finding connected components
- Solving mazes/puzzles
- Finding paths between nodes (any path, not shortest)
