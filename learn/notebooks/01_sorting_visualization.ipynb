```python
import matplotlib.pyplot as plt
import numpy as np
import time
from typing import Callable

%matplotlib inline
```

## Sorting Implementations

```python
def bubble_sort(arr):
    arr = arr.copy()
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

def selection_sort(arr):
    arr = arr.copy()
    n = len(arr)
    for i in range(n):
        min_idx = min(range(i, n), key=lambda j: arr[j])
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

def insertion_sort(arr):
    arr = arr.copy()
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

def merge_sort(arr):
    if len(arr) <= 1:
        return arr.copy()
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

def quick_sort(arr):
    if len(arr) <= 1:
        return arr.copy()
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

def python_sort(arr):
    return sorted(arr)
```

## Benchmark Function

```python
def benchmark(func, arr, repeats=3):
    times = []
    for _ in range(repeats):
        start = time.perf_counter()
        func(arr.copy())
        times.append(time.perf_counter() - start)
    return min(times) * 1000
```

## Run Benchmarks

```python
algorithms = {
    'Bubble Sort': bubble_sort,
    'Selection Sort': selection_sort,
    'Insertion Sort': insertion_sort,
    'Merge Sort': merge_sort,
    'Quick Sort': quick_sort,
    'Python sort()': python_sort,
}

sizes = [100, 500, 1000, 2000, 5000]
results = {name: [] for name in algorithms}

for size in sizes:
    arr = list(range(size))
    np.random.shuffle(arr)
    print(f"Benchmarking size {size}...")
    for name, func in algorithms.items():
        if size > 1000 and name in ['Bubble Sort', 'Selection Sort']:
            results[name].append(None)
        else:
            results[name].append(benchmark(func, arr))

print("Done!")
```

## Visualize Results

```python
plt.figure(figsize=(12, 6))

for name, times in results.items():
    valid_sizes = [s for s, t in zip(sizes, times) if t is not None]
    valid_times = [t for t in times if t is not None]
    plt.plot(valid_sizes, valid_times, marker='o', label=name)

plt.xlabel('Array Size')
plt.ylabel('Time (ms)')
plt.title('Sorting Algorithm Performance Comparison')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## Key Takeaways

1. **O(n²) algorithms** (Bubble, Selection, Insertion) are fine for small arrays but become impractical for large data
2. **O(n log n) algorithms** (Merge, Quick) scale much better
3. **Python's built-in sort** is highly optimized (Timsort algorithm)
4. For production: always use built-in sort unless you have specific requirements
